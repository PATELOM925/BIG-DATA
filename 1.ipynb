```python
import geopandas as gpd
from shapely.geometry import Point, MultiPoint
from geopy.geocoders import Nominatim

# Create a geocoder instance
geolocator = Nominatim(user_agent="geoapiExercises")

def fetch_address(coordinates):
    location = geolocator.reverse(coordinates)
    return location.address

def load_nigeria_shapefile(shapefile_path):
    return gpd.read_file(shapefile_path)


def calculate_outcircle_radius(shapefile):
    # Re-project geometries to a projected CRS (consider using a local UTM zone)
    shapefile = shapefile.to_crs('EPSG:4326')  # Example using lat-long (adjust CRS if needed)
    # Calculate centroids (handling MultiPolygon gracefully)
    shapefile["centroid"] = shapefile.geometry.apply(lambda x: x.centroid if x.geom_type == 'Polygon' else MultiPoint(x.geoms[0].centroid for geom in x.geoms))

    # Calculate outcircle radius for each boundary (consider MultiPolygon)
    def outcircle_radius_func(geom):
        if geom.geom_type == 'Polygon':
            return outcircle(geom).radius
        elif geom.geom_type == 'MultiPolygon':
            # Calculate average radius for MultiPolygons (estimate)
            radii = [outcircle(part).radius for part in geom.geoms]
            return sum(radii) / len(radii)  # Average radius
        else:
            # Handle other geometry types (if needed)
            return None

    shapefile["outcircle_radius"] = shapefile["geometry"].apply(outcircle_radius_func)

    return shapefile[['STATE', 'centroid', 'outcircle_radius', 'LGA']]  # Select relevant columns


def check_coordinates(coordinates, nigeria_shapefile):
    nigeria_bbox = nigeria_shapefile.total_bounds
    if (coordinates[0] >= nigeria_bbox[0] and coordinates[0] <= nigeria_bbox[2] and
            coordinates[1] >= nigeria_bbox[1] and coordinates[1] <= nigeria_bbox[3]):

        point = Point(coordinates)
        for index, row in nigeria_shapefile.iterrows():
            if row['geometry'].contains(point):
                return True, row['STATE'], row['LGA']
        return True, None, None  # In Nigeria, but not within a specific state
    else:
        return False, None, None  # Not in Nigeria


def check_city(coordinates, nigeria_shapefile):
    point = Point(coordinates)
    for index, row in nigeria_shapefile.iterrows():
        if row['geometry'].contains(point):
            return row['LGA']  # City name found
    return None  # Not within any city boundary
```

```python

# Example usage
shapefile_path = "/content/new_lga_nigeria_2003.shp"
nigeria_shapefile = load_nigeria_shapefile(shapefile_path)

# coordinates = (7.3775, 9.0765)  # Example coordinates
coordinates = (3.923735,7.414686)

in_nigeria, state, city = check_coordinates(coordinates, nigeria_shapefile)
if in_nigeria:
    print("Coordinates are in Nigeria.")
    if state:
        print(f"They belong to the state of {state}.")
    else:
        print("However, the specific state is unknown.")
    if city:
        print(f"They belong to the city of {city}.")
    else:
        print("However, the specific city is unknown.")
else:
    print("Coordinates are not within Nigeria.")
    address = fetch_address(coordinates)
    print(f"The address of the coordinates is: {address}")
```
